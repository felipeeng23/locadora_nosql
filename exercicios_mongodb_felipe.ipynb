```python
'''
6 - Encontre os clientes que alugaram mais de um carro.
'''
resultado = list(db.aluguel.aggregate([
    # Agrupar por cliente e contar quantos aluguéis fez
    {
        '$group': {
            '_id': '$codcliente',
            'quantidade_alugueis': { '$sum': 1 }
        }
    },
    # Filtrar apenas os que têm mais de 1
    {
        '$match': {
            'quantidade_alugueis': { '$gt': 1 }
        }
    },
    # Juntar com a coleção de clientes para pegar o nome
    {
        '$lookup': {
            'from': 'clientes',
            'localField': '_id',
            'foreignField': 'codcliente',
            'as': 'cliente_info'
        }
    },
    { '$unwind': '$cliente_info' },

    # Exibir nome e total
    {
        '$project': {
            '_id': 0,
            'nome': '$cliente_info.nome',
            'quantidade_alugueis': 1
        }
    },
    { '$sort': { 'quantidade_alugueis': -1 } }  # opcional: ordenar do maior pro menor
]))
import pandas as pd

df = pd.DataFrame(resultado)
display(df)

```

```python
'''
7 - Liste os clientes que alugaram carros mais de uma vez e os modelos e
marcas dos carros que alugaram.
'''
resultado = list(db.aluguel.aggregate([
    # Join com clientes
    {
        '$lookup': {
            'from': 'clientes',
            'localField': 'codcliente',
            'foreignField': 'codcliente',
            'as': 'cliente_info'
        }
    },
    { '$unwind': '$cliente_info' },

    # Join com carros
    {
        '$lookup': {
            'from': 'carros',
            'localField': 'codcarro',
            'foreignField': 'codcarro',
            'as': 'carro_info'
        }
    },
    { '$unwind': '$carro_info' },

    # Agrupar por cliente para contar aluguéis
    {
        '$group': {
            '_id': {
                'codcliente': '$codcliente',
                'nome': '$cliente_info.nome'
            },
            'quantidade_alugueis': { '$sum': 1 },
            'modelos_alugados': { '$addToSet': '$carro_info.modelo' },
            'marcas_alugadas': { '$addToSet': '$carro_info.marca' }
        }
    },

    # Manter apenas quem alugou mais de uma vez
    {
        '$match': {
            'quantidade_alugueis': { '$gt': 1 }
        }
    },

    # Formatar a saída
    {
        '$project': {
            '_id': 0,
            'nome': '$_id.nome',
            'quantidade_alugueis': 1,
            'modelos_alugados': 1,
            'marcas_alugadas': 1
        }
    }
]))
import pandas as pd
df = pd.DataFrame(resultado)
display(df)
```
